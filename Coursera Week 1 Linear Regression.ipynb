```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
print(plt.style.available)
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300,500])
print(f'x_train = {x_train}')
print(f'y_train = {y_train}')
```

```python
print(f"x_train.shape:{x_train.shape}")
m = len(x_train)
print(f"number of training numbers: {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"x^({i}),y^({i}) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train,y_train, marker='x',c ='red')
plt.title('Housing Prices')
plt.ylabel('price in (1000s of dollars)')
plt.xlabel('size in 1000 feets')
plt.show()
```

```python
w = 200
b = 100
print(f"w:{w}")
print(f"b:{b}")
```

```python
def compute_model_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(len(x))
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train,w,b)

plt.plot(x_train, tmp_f_wb, c='red', label = 'Our Prediction')
plt.scatter(x_train, y_train, marker ='x', c ='blue', label='Actual Values')
plt.title('Housing Prices')
plt.ylabel('price')
plt.xlabel('feet')
plt.legend(loc='best')
plt.show()
```

```python
x = np.array([1.2])
compute_model_output(x,200,100)
```
